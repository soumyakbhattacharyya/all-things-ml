## Matplotlib

Reference - https://matplotlib.org/gallery/index.html

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline 
#this will help seeing the plot inside jupyter notebook; alternative is plt.show()
```

### Step 1 - simple plotting example

```python
# create an ndarray by using linspace function
start = 1
stop = 10
numberOfElements = 20
x_axis = np.linspace(start, stop, numberOfElements)
print('x_axis is ', x_axis)
y_axis = x_axis ** 2
print('y_axis is ', y_axis)
```

there are 2 methods, 1) functional and 2) object oriented method to plot

```python
# functional plot
plt.plot(x_axis, y_axis) # 
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('Simple Plot')
```

#### Simple Multi Plot

```python
# lets create two sections
numOfRows = 1
numOfCols = 2
plt.subplot(numOfRows,numOfCols,1)

# plotting first plot
plt.plot(y_axis, y_axis, 'r') # r signifies red colour
plt.xlabel('x label') 
plt.ylabel('y label')
plt.title('1st Plot')

plt.subplot(numOfRows,numOfCols,2)

# plotting second plot
plt.plot(x_axis, y_axis)
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('2nd Plot')
```

### Step 2 - object oriented plotting

```python
fig = plt.figure()

# create canvas
left = 0.1
bottom = 0.1
width = 0.9
height = 0.9 

axes = fig.add_axes([left,bottom, width, height])
axes.plot(x_axis, y_axis)
axes.set_xlabel('x')
axes.set_ylabel('y')
```

Demystifying `plt.subplot(), plt.subplots(), plt.axes(), fig.add_subplot(), fig.subplots(), and fig.add_axes()`. Note first three functions are for functional type plotting and last 3 are for object oriented plotting.

```python
# Create a figure
fig = plt.figure()

# Add a subplot
ax = fig.add_subplot()

# Equivalent method
ax = fig.add_subplot(111) # 1 axes in a 1 row by 1 column axes grid

# Another equivalent but more general method
ax = fig.add_subplot(1, 1, 1)
```

```python
# Create a figure
fig2 = plt.figure()

# Create a subplot
ax = fig2.subplots()
```

```python
# Create a figure and subplot simlutaneously
fig3, ax = plt.subplots(2, 3)
plt.tight_layout()
```

plt.subplot() returns only one axes and create a figure object automatically. You can change figure-level attributes or save the figure by calling ax.figure.xxx() instead.

`Figure` with `multiple axes`

```python
fig5 = plt.figure()
ax1 = fig5.add_subplot(221)
ax2 = fig5.add_subplot(222)
ax3 = fig5.add_subplot(223)
ax4 = fig5.add_subplot(224)

ax1.annotate('I am ax1',(.5,.5),xycoords='axes fraction', va='center',ha='center')
ax2.annotate('I am ax2',(.5,.5),xycoords='axes fraction', va='center',ha='center')
ax3.annotate('I am ax3',(.5,.5),xycoords='axes fraction', va='center',ha='center')
ax4.annotate('I am ax4',(.5,.5),xycoords='axes fraction', va='center',ha='center')

plt.tight_layout()
```

```python
fig6 = plt.figure()
ax1 = fig6.add_axes([0,.6,.5,.5])
ax2 = fig6.add_axes([.6,.6,.5,.5])
ax3 = fig6.add_axes([0.6,0,.5,.5])
ax4 = fig6.add_axes([0,0,.5,.5])

ax1.annotate('I am ax1',(.5,.5),xycoords='axes fraction', va='center',ha='center')
ax2.annotate('I am ax2',(.5,.5),xycoords='axes fraction', va='center',ha='center')
ax3.annotate('I am ax3',(.5,.5),xycoords='axes fraction', va='center',ha='center')
ax4.annotate('I am ax4',(.5,.5),xycoords='axes fraction', va='center',ha='center')

plt.tight_layout()
```

Using axes

```python
fig7 = plt.figure()
ax1 = fig7.add_axes([0,.6,.5,.5])
ax1.plot(x_axis,y_axis)
ax2 = fig7.add_axes([.6,.6,.5,.5])
ax2.plot(y_axis,x_axis)
ax3 = fig7.add_axes([0.6,0,.5,.5])
ax3.plot(y_axis ** 7,x_axis)
ax4 = fig7.add_axes([0,0,.5,.5])
ax4.plot(x_axis ** 8,y_axis)

ax1.annotate('I am ax1',(.5,.5),xycoords='axes fraction', va='center',ha='center')
ax2.annotate('I am ax2',(.5,.5),xycoords='axes fraction', va='center',ha='center')
ax3.annotate('I am ax3',(.5,.5),xycoords='axes fraction', va='center',ha='center')
ax4.annotate('I am ax4',(.5,.5),xycoords='axes fraction', va='center',ha='center')
```

### Figure Size, Legend and Label

```python
fig10, axes = plt.subplots(nrows=2,ncols=1,figsize=(10,10))
axes[0].plot(x_axis,y_axis, label='abc')
axes[1].plot(y_axis,x_axis)
axes[0].set_title('Fancy Curve')
axes[0].legend()
plt.tight_layout()
```

### Tweaking attributes of a graph

```python
z = np.linspace(8,90,50)
print(z)

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(z ** 3, color='green', linewidth=4, alpha=.6, linestyle="--", marker="o", markerfacecolor='red', markersize=8)
```

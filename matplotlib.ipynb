## Matplotlib

Reference - https://matplotlib.org/gallery/index.html

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline 
#this will help seeing the plot inside jupyter notebook; alternative is plt.show()
```

### Step 1 - simple plotting example

```python
# create an ndarray by using linspace function
start = 1
stop = 10
numberOfElements = 20
x_axis = np.linspace(start, stop, numberOfElements)
print('x_axis is ', x_axis)
y_axis = x_axis ** 2
print('y_axis is ', y_axis)
```

there are 2 methods, 1) functional and 2) object oriented method to plot

```python
# functional plot
plt.plot(x_axis, y_axis)
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('Simple Plot')
```

#### Simple Multi Plot

```python
# lets create two sections
numOfRows = 1
numOfCols = 2
plt.subplot(numOfRows,numOfCols,1)

plt.plot(y_axis, y_axis, 'r') # r signifies red colour
plt.xlabel('x label') 
plt.ylabel('y label')
plt.title('1st Plot')

plt.subplot(numOfRows,numOfCols,2)

plt.plot(x_axis, y_axis)
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('2nd Plot')
```
